```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_california_housing

house_price_dataset = fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a pandas dataframe
import pandas as pd

house_price_dataframe = pd.DataFrame(
    house_price_dataset.data,
    columns=house_price_dataset.feature_names
)

house_price_dataframe.head()
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe['Price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
#is theere any missing values in th dataset
house_price_dataframe.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

```python
#understanding the correlation between the various features in the dataset
# 1. positive correlation
# 2. negative correlation

correlation = house_price_dataframe.corr()
```

```python
# constructing the heatmap to understand the correlation
plt.figure(figsize=(14,14))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#splitting the data and target
X = house_price_dataframe.drop('Price', axis=1)
Y = house_price_dataframe['Price']
```

```python
print(X)
print(Y)
```

```python
#spilitting the dat into training and test data
X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# traing the model with X_train
from xgboost import XGBRegressor

model = XGBRegressor(random_state=2)

model.fit(X_train, Y_train)
```

```python
# Model Training
# XGBoost Regressor
# loading the model
model = XGBRegressor()
```

```python
print(X_train.shape)
print(Y_train.shape)

print(X_test.shape)
print(Y_test.shape)
```

```python
# Evaluation
# prediction on training data
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
```

```python
print("R squared error : ", score_1)
print("Mean Absolute Error : ", score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Prediction on test data


```python
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("R squared error : ", score_1)
print("Mean Absolute Error : ", score_2)
```
